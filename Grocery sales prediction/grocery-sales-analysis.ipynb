# Predicting net sales using linear regression with multiple independent variable

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
```

```python
X=file[['Quantity', 'Original Price (Rs.)', 'Discount', 'Total sales', 'Total Sales Dis']]
```

```python
Y=file['Net Sales']
```

```python
X
```

```python
# splitting 70% of the data into train, 30% test split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
reg_m=LinearRegression()
#train data
reg_m.fit(X_train, Y_train)
```

```python
# r^2 accuracy level
reg_m.score(X_train,Y_train)
```

```python
#using testing data to check accuracy of the model
y_pred=reg_m.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
#plotting the actual and predicted value
plt.scatter(Y_test,y_pred)
plt.xlabel('Y_test')
plt.ylabel('y_pred')
```

```python
#calculating the model error
mse = metrics.mean_squared_error(Y_test, y_pred)
print("Mean Squared Error:",mse)
```

```python
print('MAE:', metrics.mean_absolute_error(Y_test, y_pred))
print('MSE:', metrics.mean_squared_error(Y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))
```
